```python
# import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import scipy.stats as stats
from scipy import io
import seaborn as sns
sns.set(color_codes=True)
plt.rcParams["patch.force_edgecolor"] = True
%matplotlib inline
```

<h3>Pre-Lab: </h3>

Degrees of freedom = N-2 = 20.  

At 95% confidence, we <b>cannot reject the null hypothesis</b>  
  
At 99% confidence interval, we can <b>reject the null hypothesis</b>

```python
# Import data

df = pd.read_csv("CityData_VerminLab.csv")

roach = np.array(df["cockroach reports"])
rodent = np.array(df["rodent reports"])
temp = np.array(df["mean temperature (C)"])
```

<h3>(a) Scatter plots</h3>

```python
# Initialize figure
fig = plt.figure(figsize=(12,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

# Scatter plot
ax1.plot(temp, roach, 'bo')

ax2.plot(temp, rodent, 'ro')

# Customize plot
ax1.set_title("Scatter Plot: Cockroach")
ax1.set_xlabel("Mean Annual Temperature (C)")
ax1.set_ylabel("Percentage of household")

ax2.set_title("Scatter Plot: Rodents")
ax2.set_xlabel("Mean Annual Temperature (C)")
ax2.set_ylabel("Percentage of household")
               
plt.show()
```

We can see a strong positive correlation between mean annual temperature and percentage of household infested with roaches. We also see a moderate negative correlation between mean annual temperature and percentage of household infested with rodents.

<h3>(b) Correlation coefficients</h3>

```python
rho_roach = np.corrcoef(temp, roach)[1,0]
rho_rodent = np.corrcoef(temp, rodent)[1,0]

print("The correlation coefficient of temp vs roach is  : %.3f" %rho_roach)
print("The correlation coefficient of temp vs rodent is : %.3f" %rho_rodent)
```

<h3>(c) t-statistic</h3>

```python
# T-VALUE FUNCTION

t_statistic = lambda rho,N:rho*np.sqrt(N-2)/np.sqrt(1-(rho**2))
```

```python
N = len(roach)

t_roach = t_statistic(rho_roach,N)
t_rodent = t_statistic(rho_rodent,N)

print("The t-statistic of roach vs. temperature is : %.3f" %t_roach)
print("The t-statistic of rodent vs. temperature is: %.3f" %t_rodent)
```

At 95% and 99% confidence, for 23 degrees of freedom, the null hypothesis should be rejected for both roaches and rodents since we have a strong correlation.   
  
We can conclude that we are 99% confident that this correlation is not purely by chance and has a significance.

<h3>(d) Bootstrap resampling</h3>

```python
N = len(roach)
N_steps = 10000
N_subsample = 100

# Correlation coefficient array
resampled_roach_rho = np.zeros((N_steps))
resampled_rodent_rho = np.zeros((N_steps))

sub_roaches = np.zeros((N_subsample))
sub_rodents = np.zeros((N_subsample))
sub_temp = np.zeros((N_subsample))

for i in range(N_steps):
    # Shuffle the indices
    resample_i = np.floor(np.random.rand(N_subsample)*len(temp)).astype(int)
    
    # Resample values
    sub_roaches = roach[resample_i]
    sub_temp = temp[resample_i]
    sub_rodents = rodent[resample_i]
    
    # Correlation coefficient array
    resampled_roach_rho[i] = np.corrcoef(sub_temp, sub_roaches)[1,0]
    resampled_rodent_rho[i] = np.corrcoef(sub_temp, sub_rodents)[1,0]

##################
# Plot histogram #
##################

# Initialize figure
fig = plt.figure(figsize=(12,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.hist(resampled_roach_rho, bins=50)

# Customize plot
ax1.axvline(resampled_roach_rho.mean(), color='r', linestyle='dashed', linewidth=2,\
           label="Mean")
ax1.set_title("Histogram of resampled roaches")
ax1.set_xlabel("Resampled correlation coefficient")
ax1.set_ylabel("Frequency")
ax1.legend()

ax2.hist(resampled_rodent_rho, bins=50)

# Customize plot
ax2.axvline(resampled_rodent_rho.mean(), color='r', linestyle='dashed', linewidth=2,\
           label="Mean")
ax2.set_title("Histogram of resampled rodents")
ax2.set_xlabel("Resampled correlation coefficient")
ax2.set_ylabel("Frequency")
ax2.legend()

plt.show()
```

<h3>(e) Mean and std. of correlation coefficient</h3>

```python
mean_rhoBS_roach = np.mean(resampled_roach_rho)
std_rhoBS_roach = np.std(resampled_roach_rho)
sem_rhoBS_roach = stats.sem(resampled_roach_rho)
#rhoBS_95_roach = stats.norm.interval(0.95, loc = mean_rhoBS_roach, scale = std_rhoBS_roach)

mean_rhoBS_rodent = np.mean(resampled_rodent_rho)
std_rhoBS_rodent = np.std(resampled_rodent_rho)
sem_rhoBS_rodent = stats.sem(resampled_rodent_rho)
#rhoBS_95_rodent = stats.norm.interval(0.95, loc = mean_rhoBS_rodent, scale = std_rhoBS_rodent)


print("Bootstrap resampling for roach:\n")
print("\tMean rho                   = %.3f" %mean_rhoBS_roach)
print("\tStandard deviation of rho  = %.3f" %std_rhoBS_roach)
print("\tStandard error of mean rho = %.3f" %sem_rhoBS_roach)
#print("\t95 percent confidence interval of rho = %.3f to %.3f" %rhoBS_95_roach)
print("\tThe t-value of rho: %.3f" %t_statistic(mean_rhoBS_roach,N))

print("\n\nBootstrap resampling for rodent:\n")
print("\tMean rho                   = %.3f" %mean_rhoBS_rodent)
print("\tStandard deviation of rho  = %.3f" %std_rhoBS_rodent)
print("\tStandard error of mean rho = %.3f" %sem_rhoBS_rodent)
#print("\t95 percent confidence interval of rho = %.3f to %.3f" %rhoBS_95_rodent)
print("\tThe t-value of rho: %.3f" %t_statistic(mean_rhoBS_rodent,N))
```

<h3>(f)</h3>

After bootstrap resampling, we see that the t-values are less than what we found out in part (c). Therefore, at 95% confidence, we can reject the null hypothesis but with less confidence than part (c).

<h3>(g) Jackknifing</h3>

```python
# Index for jackknifing: 0 to 24 cities
idx = np.arange(0,N).astype(int)

sub_temp = np.zeros((N-1))
sub_roaches = np.zeros((N-1))
sub_rodents = np.zeros((N-1))

# pseudo values
rho_roach_sv = np.zeros((N))
rho_rodent_sv = np.zeros((N))

for i in range(N):
    # Filter each element from array iteratively
    idx_iter= list(filter(lambda x: x!=i,idx))
    
    sub_temp = temp[idx_iter]
    sub_roaches = roach[idx_iter]
    sub_rodents = rodent[idx_iter]
    
    # Correlation coefficient
    rho_roach_i = np.corrcoef(sub_temp, sub_roaches)[1,0]
    rho_rodent_i = np.corrcoef(sub_temp, sub_rodents)[1,0]
    
    rho_roach_sv[i] = (N*rho_roach) - (N-1)*rho_roach_i
    rho_rodent_sv[i] = (N*rho_rodent) - (N-1)*rho_rodent_i
    
print("The Jackknife solution for roach is %.3f +/- %.3f" \
     %(np.mean(rho_roach_sv), stats.sem(rho_roach_sv)))

print("\nThe Jackknife solution for rodent is %.3f +/- %.3f" \
     %(np.mean(rho_rodent_sv), stats.sem(rho_rodent_sv)))
```

The mean in part (e) and (g) are almost same. The standard error on mean is negligible (almost 0) for bootstrapping while it is significant for Jackknifing. The reason is that we are bootstrapping a very large number of times, therefore reducing the standard error significantly.
