<h3>(b) Function to compute one way travel time</h3>

```python
# N = station locations [x,y] in km
# loc = possible Yeti location [x,y] in km
# v = speed of sound
def travel_time(N, loc, v):
    "compute one way travel time"
    t = np.zeros((len(N[1,:])))
    for i in range(len(N[1,:])):
        t[i]= (np.sqrt(np.power((loc[0]-N[0,i]),2) + np.power((loc[1]-N[1,i]),2)))/v
    return t;
```

```python
# test the function
v = 0.301 # speed of sound
print(travel_time(N, [5,15], v))
```

<h3>(c) Trial models</h3>

```python
# Number of trial models
n = 1000000

# Initial random number based on the range
x_trial = np.random.uniform(10,30,n)
y_trial = np.random.uniform(-15,25,n)

trial_loc = np.array([x_trial, y_trial])

# Plot trial models
# Plot
fig = plt.figure(figsize=(12,10))

ax1 = fig.add_subplot(221)
#ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax3.scatter(x_trial, y_trial, c = 'b', marker='.', s=5, alpha=0.1)

# Plot station locations
for i in range(len(x)):
    ax3.plot(x[i],y[i], label='Station '+str(i+1), marker='*', markersize=15)

ax3.legend(loc='lower left')
ax3.set_xlabel('x (km)')
ax3.set_ylabel('y (km)')
ax3.set_title('Search space: 10^6 trial models')


# Histogram
ax1.hist(x_trial, range=(np.amin(x), np.amax(x)), bins=100, color='blue', edgecolor='none', alpha=0.5)
ax1.set_ylabel('Frequency')

ax4.hist(y_trial, range=(np.amin(y_trial),np.amax(y_trial)), bins=100, color='blue',\
         orientation='horizontal', edgecolor='none', alpha=0.5)
ax4.set_xlabel('Frequency')
    
plt.show()
```

<h3>(d) </h3>

L-1 norm will be a better estimate for the misfit due to the way travel times are recorded. Laplace distribution is more appropriate when the uncertainties are within a very narrow window.

<h3>(e) Calculate misfit</h3>

```python
# Predicted travel-times 
t_predicted = np.zeros((5, len(x_trial)))

# Misfit for 1 observation
residual = np.zeros((5, len(x_trial)))

for i in range(len(x_trial)):
    t_predicted[:,i] = travel_time(N, [trial_loc[0,i], trial_loc[1,i]], v)
    residual[:,i] = np.abs(t_predicted[:,i] - t_obs)/ t_sigma
    
# L1 misfit
L1_misfit = np.sum(residual, axis=0)
```

<h3>(f) Likelihood of each trial: Posterior samples</h3>

```python
# Laplacian likelihood
L = np.exp(-np.sqrt(2)*L1_misfit)

# Likelihood/ maximum likelihood
l = L/np.amax(L)
#posterior_temp = np.zeros(n)

# Posterior model index 
x_post = np.zeros(n)
y_post = np.zeros(n)

# Accept posterior models
for j in range(n):
    if l[j] > np.random.uniform():
        x_post[j] = x_trial[j]
        y_post[j] = y_trial[j]
        

        
# Plot
fig = plt.figure(figsize=(12,10))

ax1 = fig.add_subplot(221)
#ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

# Plot station locations
for i in range(len(x)):
    ax3.plot(x[i],y[i], label='Station '+str(i+1), marker='*', markersize=15)

ax3.legend(loc='lower left')
ax3.set_xlabel('x (km)')
ax3.set_ylabel('y (km)')
ax3.set_title('Posterior models')

# Plot posterior models
ax3.scatter(x_post, y_post, c = 'blue', marker='.', s=5, alpha=0.4)

# Posterior model estimates 
x_post = list(filter(lambda x: x!=0, x_post))
y_post = list(filter(lambda x: x!=0, y_post))

# Histogram
ax1.hist(x_post, range=(np.amin(x), np.amax(x)), bins=240, color='blue',\
         edgecolor='none', alpha=0.5)

ax1.set_ylabel('Frequency')

ax4.hist(y_post, range=(0,20), bins=100, color='blue',\
         orientation='horizontal', edgecolor='none', alpha=0.5)

ax4.set_xlabel('Frequency')

plt.show()
```

<h3>(g)</h3>

```python
corrcoef = np.corrcoef(x_post, y_post)[0][1]

print("Most likely location of Yeti: x = %.2f km, y = %.2f km" %(np.mean(x_post), np.mean(y_post)))
print("\nUncertainty in the model parameters: x = %.2f km, y = %.2f km" %(stats.sem(x_post), stats.sem(y_post)))
print("\nThe correlation coefficient between the model parameters = %.2f" %corrcoef)
```

<h3>(h) Posterior travel time and uncertainties in the Yecho recordings</h3>

```python
#
t_post_predicted = np.zeros((5, len(x_post)))

for i in range(len(x_post)):
    t_post_predicted[:,i] = travel_time(N, [x_post[i], y_post[i]], v)

most_likely_post = np.mean(t_post_predicted, axis=1)
uncertainty_post = stats.sem(t_post_predicted, axis=1)

print("Posterior travel times for each station respectively:", np.round(most_likely_post,2) )
print("Estimated uncertainties:                             ", np.round(uncertainty_post,2) )
```
