<h3>1. Propagating vs. non propagating waves</h3>

General form of wave in a rod (propagating + non propagating):

$$y(x,t) = Y_0 e^{i(\omega t - kx)} + Y_1 e^{-kx + i\omega t}$$

Since the rod is semi-infinite, no waves travel in the negative direction.

Here, $k = \sqrt{\frac{\omega}{a}}$ and $a^2 = \frac{EI}{\rho A}$. Therefore, $a^2 = 2.137$ and $k = 65.6 m^{-1}$.

Given:
    $$
    \begin{aligned}
    &E = 200 GPa;\\ 
    &\rho = 7800 kg/m^3;\\
    &A = 1 mm^2 = 10^{-6} m^2;\\ 
    &I = A^2 / 12;\\
    &\omega = 2\pi (1kHz) = 6283.2\ Hz;\\ 
    &F_0 = 1N;\\ 
    &F = F_0e^{i\omega t};
    \end{aligned}
    $$

<b>Boundary condition:</b>  

at x = 0:

1. Bending moment $M_b = 0$
    $$
    \begin{aligned}
    &\frac{\partial^2y}{\partial x^2} = 0
    \\
    &Y_0 - Y_1 = 0 
    \\
    &Y_0 = Y_1 
    \end{aligned}
    $$ 
    
2. Shear force = x-derivative of bending moment
    $$
    \begin{aligned}
    \\
    &V = \frac{\partial M_b}{\partial x}
    \\
    &V = EI\frac{\partial^3y}{\partial x^3}
    \\
    &F_0e^{i\omega t} = -EI\frac{\partial^3y}{\partial x^3}
    \\
    &-iY_0 + Y_1 = \frac{F_0}{EIk^3}
    \end{aligned}
    $$
  
  
Solving the two equations of the boundary condition, we get:
    $$
    Y_0 = Y_1 = \frac{F_0 (1 + i)}{2 EIk^3}
    $$
    
Plugging the constant values, we get: $\frac{F_0}{2 EIk^3} = 5.31 \times 10^{-5} m$.

<h4>(a) Propagating wave:</h4>
  
$$
\begin{aligned}
&y_p(x,t) = Y_0e^{i(wt-kx)}
\\
&y_p(x,t) = \frac{F_0 (1 + i)}{2 EIk^3} e^{i(wt-kx)}
\\
&y_p(x,t) = 5.3 \times 10^{-5} e^{i(6283.2t - 65.6x)} (1+i)
\end{aligned}
$$

<h4>(b) Wavelength of the propagating wave</h4>
  
  
$$
\lambda = \frac{2\pi}{k}
\\
\lambda = 9.578\ cm
$$

<h4>(b) Decaying wave:</h4>
  
$$
\begin{aligned}
&y_np(x,t) = Y_1 e^{-kx + i\omega t}
\\
&y_np(x,t) = \frac{F_0 (1 + i)}{2 EIk^3} e^{-kx + i\omega t}
\\
&y_np(x,t) = 5.3 \times 10^{-5} e^{-65.6x + 6283.2it} (1 + i)
\end{aligned}
$$

<h4>(c) Plot real part of displacement</h4>
  
  
$$
Re(y_p(x,t=0)) = 7.49 \times 10^{-5} [cos(\frac{\pi}{4} - 65.6x)]
\\
Re(y_np(x,t=0)) = 5.3 \times 10^{-5} \times e^{-65.6x}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

A = 7.49e-5
B = 5.3e-5
k = 65.6
lamb = 0.09578

x = np.linspace(0,5*lamb,num=1000) 
y_p = A*np.cos((np.pi/4) - k*x)
y_np = B*np.exp(-k*x)
```

```python
fig = plt.figure(figsize=(11,5))
ax = fig.add_subplot(111)

ax.plot(x, y_p*1e5, 'r', label='propagating')
ax.plot(x,y_np*1e5, 'b', label = 'decaying')
ax.set_title('Physical displacement at t=0')
ax.set_xlabel('x (m)')
ax.set_ylabel('y (10^-5 m)')
plt.legend()
plt.show()
```

<h3>2. $k-\omega$ diagram</h3>

Thin steel rod. 
  
Given:
    $$
    \begin{aligned}
    &E = 200 GPa;\\ 
    &\rho = 7800 kg/m^3;\\
    &G = 160 GPa;\\
    &A = 1 mm^2 = 10^{-6} m^2;\\ 
    &I = A^2 / 12;\\
    &\kappa = 5/6;\\ 
    \end{aligned}
    $$


Timoshenko's wave equations:

$$
GA\kappa \big( \frac{\partial \psi}{\partial x} - \frac{\partial^2y}{\partial x^2} \big) + \rho A\frac{\partial^2y}{\partial t^2} = 0
\\
\\
GA\kappa \big( \frac{\partial y }{\partial x} - \psi\big) + EI\frac{\partial^2\psi}{\partial x^2} = \rho I\frac{\partial^2\psi}{\partial t^2}
$$

<h4>(a)</h4>
Harmonic propagation:
$$
y(x,t) = Y_0e^{i(\omega t - kx)}\\
\psi (x,t) = \psi_0e^{i(\omega t - kx)}
$$
  
$$
\frac{\partial \psi}{\partial x} = -ik\psi_0e^{i(\omega t - kx)};
\frac{\partial \psi}{\partial t} = i\omega\psi_0e^{i(\omega t - kx)}
\\
\frac{\partial^2 \psi}{\partial x^2} = -k^2\psi_0e^{i(\omega t - kx)};
\frac{\partial^2 \psi}{\partial t^2} = -\omega^2\psi_0e^{i(\omega t - kx)}
\\
\frac{\partial y}{\partial x} = -ikY_0e^{i(\omega t - kx)};
\frac{\partial y}{\partial t} = i\omega Y_0e^{i(\omega t - kx)}
\\
\frac{\partial^2 y}{\partial x^2} = -k^2Y_0e^{i(\omega t - kx)};
\frac{\partial^2 y}{\partial x^2} = -\omega^2Y_0e^{i(\omega t - kx)}
$$

Plugging in the above wave equation:
$$
GA\kappa [-ik\psi_0 + k^2Y_0] + \rho A\omega^2Y_0 = 0
\\
GA\kappa [-ikY_0 - \psi_0] + EI[-k^2\psi_0] = -\rho I\omega^2\psi_0
$$
  
  
Solving the two equations, we get:
$$
(EIGA\kappa)k^4 - \rho IA(G\kappa + E)k^2\omega^2 + (\rho IA^2)\omega^4 - (\rho A^2G\kappa)\omega^2 = 0
$$
  
  
We see that the above equation is of the given form with values:
$$
m = \frac{\rho (G\kappa + E)}{EG\kappa} = 9.75 \times 10^{-7}
\\
n = \frac{\rho A}{EG\kappa} = 2.90 \times 10^{-24}
\\
p = \frac{\rho A}{EI} = 0.47 
$$

<h4>(b) Plot Timoshenko and Bernoulli-Euler theories</h4>

```python
# Bernoulli-Euler
a = 1.46
omega = 2*np.pi*(np.linspace(1,5e6,num=10000))
k_BE = np.sqrt(omega/a)

# Timoshenko
m = 9.75*1e-7
n = 2.9*1e-24
p = 0.47

temp1 = m*(omega**2)
temp2 = ((m**2)*np.power(omega,4)) - (4*n*np.power(omega,4))\
        + (4*p*(omega**2))
k_TS = np.sqrt((np.sqrt(temp2) - temp1)/2)

fig = plt.figure(figsize=(9,5))
plt.plot(k_BE, omega/1e6, 'r', label='Bernoulli-Euler')
plt.plot(-k_BE, omega/1e6, 'r')
plt.plot(k_TS, omega/1e6, 'b', label='Timoshenko')
plt.plot(-k_TS, omega/1e6, 'b')
plt.xlabel('k (m^-1)')
plt.ylabel('omega(2*pi x MHz)')
plt.legend()
plt.show()
```

<h4>(c) Group velocity vs frequency</h4>
  
Bernoulli Euler group velocity:
$$C^{BE}_g = 2ak$$
  
Timoshenko group velocity:
$$C^{TS}_g = \frac{2k^3 - mk\omega^2}{mk^2\omega - n\omega^3 + p\omega}$$

```python
cg_BE = 2*a*k_BE

cg_TS = ((2*np.power(k_TS,3) - (m*np.multiply(k_TS,omega**2)))/\
         ((m*np.multiply(k_TS**2,omega)) - (n*omega**3) + (p*omega)))

plt.plot(cg_BE, omega/1e6, 'r', label='Bernoulli-Euler')
plt.plot(cg_TS, omega/1e6, 'b', label='Timoshenko')
plt.xlabel('Group velocity (m/s)')
plt.ylabel('omega(2*pi MHz)')
plt.legend()
plt.show()
```

Bernoulli Euler has nonlinear group velocity increase, and Timoshenko group velocity is negative for larger frequencies suggesting that envelop of waves travel in backward direction.

<h4>(d)</h4>

```python
E = 2e11
rho = 7800
v_compressional = np.sqrt(E/rho)

# We can use this algorithm because the group velocity
# of timoshenko is monotonically increasing
for i in range(len(cg_TS)):
    if v_compressional < -cg_TS[i]:
        break
print ("The compressional wave velocity is %.2f m/s" %v_compressional)
print("The flexural wave group velocity is %.2f m/s" %-cg_TS[i])
print("The frequency when they are equal is %.2f MHz" %(omega[i]/(2e6*np.pi)))
```
